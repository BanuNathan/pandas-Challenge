```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_1 = pd.read_csv(school_data_to_load)
student_data_1= pd.read_csv(student_data_to_load)
school_data=school_data_1.copy()
student_data=student_data_1.copy()
student_data["math_score"]=student_data.loc[:, "math_score"].astype(float)
student_data["reading_score"]=student_data.loc[:, "reading_score"].astype(float)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

```python
# Drop all rows with missing information
school_data_complete = school_data_complete.dropna(how='any')
```

```python
scl_cnts=school_data_complete["school_name"].value_counts()
```

```python
#grouping the Data by school
grouped_school_df = school_data_complete.groupby(['school_name'])
```

```python
# extracting total students for each schoolin a dataframe
total_students_per_school = grouped_school_df["size"].mean()

total_students_per_school_df=pd.DataFrame(total_students_per_school)

total_students=total_students_per_school_df["size"].sum()
```

```python
#school_summary_test=pd.DataFrame({"size":scl_cnts,"total_students":total_students_per_school})
#school_summary_test.head()
```

```python
# extracting budget for each schoolin a dataframe
budget_by_school = grouped_school_df["budget"].mean() 
budget_by_school_df=pd.DataFrame(budget_by_school)
```

```python
# extracting budget pet student for each schoolin a dataframe
budget_per_student = grouped_school_df["budget"].mean()/grouped_school_df["size"].unique()
#budget_per_student_df=pd.DataFrame(budget_per_student)
budget_per_student_df=pd.DataFrame({"Budget per student":budget_per_student})
```

```python
# extracting average math score for each schoolin a dataframe
average_math_by_school = grouped_school_df["math_score"].mean()
average_math_by_school_df=pd.DataFrame(average_math_by_school)
```

```python
# extracting average reading score for each schoolin a dataframe
average_reading_by_school = grouped_school_df["reading_score"].mean()
average_reading_by_school_df=pd.DataFrame(average_reading_by_school)
```

```python
#extracting type of scholl into a data frame
type_of_school = grouped_school_df["type"].unique()
type_of_school_df=pd.DataFrame(type_of_school)
```

```python
math_passed = school_data_complete.loc[school_data_complete.math_score >= 70]
math_passed_df=pd.DataFrame(math_passed)
math_grp_df = math_passed_df.groupby(['school_name'])
```

```python
#extracting percentage math passed in each school
percent_math=(math_grp_df["student_name"].count()/math_grp_df["size"].unique())*100
percent_math_df=pd.DataFrame({"Percentage math":percent_math})
percent_math_df["Percentage math"]=percent_math_df.loc[:, "Percentage math"].astype(float)
```

```python
#group by passed reading
read_passed = school_data_complete.loc[school_data_complete.reading_score >= 70]
read_passed_df=pd.DataFrame(read_passed)
read_grp_df = read_passed_df.groupby(['school_name'])
```

```python
#reading percent by school
percent_read=(read_grp_df["student_name"].count()/read_grp_df["size"].unique())*100
percent_read_df=pd.DataFrame({"Percentage read":percent_read})
#percent_read_df=pd.DataFrame(percent_read)
percent_read_df["Percentage read"]=percent_read_df.loc[:, "Percentage read"].astype(float)
```

```python
read_math_passed= school_data_complete.loc[(school_data_complete.reading_score >= 70) & (school_data_complete.math_score >= 70)]
read_math_passed_df=pd.DataFrame(read_math_passed)
read_math_passed_df
read_math_grp_df = read_math_passed_df.groupby(['school_name'])
```

```python
#reading percent by school
percent_read_math=(read_math_grp_df["student_name"].count()/read_math_grp_df["size"].unique())*100
#percent_read_math_df=pd.DataFrame(percent_read_math)
percent_read_math_df=pd.DataFrame({"Percentage OverAll Passed":percent_math})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
lessthan600=school_summary_final["Budget per student"]<600
lessthan_600_df=school_summary_final.loc[lessthan600]
lessthan_600_df.insert(loc=4,column='avg_bud',value='lessthan600')
lessthan_600_df

bet_600_610=(school_summary_final["Budget per student"]>=600) & (school_summary_final["Budget per student"]<620)
bet_600_610_df=school_summary_final.loc[bet_600_610]
bet_600_610_df.insert(loc=4,column='avg_bud',value='bet_600_610')
bet_600_610_df

bet_610_640=(school_summary_final["Budget per student"]>=620) & (school_summary_final["Budget per student"]<640)
bet_610_640_df=school_summary_final.loc[bet_610_640]
bet_610_640_df.insert(loc=4,column='avg_bud',value='bet_610_640')
bet_610_640_df

morethan640=school_summary_final["Budget per student"]>=640
morethan640_df=school_summary_final.loc[morethan640]
morethan640_df.insert(loc=4,column='avg_bud',value='morethan640')
morethan640_df

#result = pd.merge(lessthan_600_df,bet_600_610_df,how='outer')
vertical_stack = pd.concat([lessthan_600_df, bet_600_610_df,bet_610_640_df,morethan640_df], axis=0)
vertical_stack_group=  vertical_stack.groupby(['avg_bud'])
vertical_stack_group.count()
```

```python
school_summary_final["Percentage math"]=school_summary_final.loc[:, "Percentage math"].astype(float)
school_summary_final["Percentage read"]=school_summary_final.loc[:, "Percentage read"].astype(float)
school_summary_final["Percentage OverAll Passed"]=school_summary_final.loc[:, "Percentage OverAll Passed"].astype(float)
school_summary_final["Budget per student"]=school_summary_final.loc[:, "Budget per student"].astype(float)
```

```python
table1 = vertical_stack_group["reading_score"].mean()
table1_df=pd.DataFrame(table1)
table1_df

table2 = vertical_stack_group["math_score"].mean()
table2_df=pd.DataFrame(table2)
table2_df

table3 = vertical_stack_group["Percentage math"].sum()/vertical_stack_group["Percentage math"].count()
table3_df=pd.DataFrame(table3)
table3_df

table4 = vertical_stack_group["Percentage read"].sum()/vertical_stack_group["Percentage read"].count()
table4_df=pd.DataFrame(table4)
table4_df

table5 = vertical_stack_group["Percentage OverAll Passed"].sum()/vertical_stack_group["Percentage OverAll Passed"].count()
table5_df=pd.DataFrame(table5)
table5_df

a=pd.merge(table1,table2,left_index=True,right_on="avg_bud")
b=pd.merge(a,table3,left_index=True,right_on="avg_bud")
c=pd.merge(b,table4,left_index=True,right_on="avg_bud")
budget_summary=pd.merge(c,table5,left_index=True,right_on="avg_bud")
budget_summary
```
